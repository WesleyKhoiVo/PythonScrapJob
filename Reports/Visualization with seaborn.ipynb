The Data jobs are kinda friendly for people who are new to it, since jobs that require 0-1 year experience is not low(most of them are Data Analyst jobs).
Move on to 2-3 years experience, the need for them is high. Usually, at around 2-3 years experience, Data Analysts are good enough so company don't have to train them as much as the 0-1 year experience. At 3 years experience jobs, there are some Data Scientist and Data Engineer jobs here.
Jobs at 4 years experience are lower because 3-4 years are pratically not much different, so most of the jobs posting required 3 years is normal.
However, 5 years experience jobs are highly demanded because these people are experts in Data Science. Every company needs them.
