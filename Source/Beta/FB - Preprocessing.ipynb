```python
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('FB-BD.csv')
```

```python
kw = ['lương', 'việc', 'offer', 'salary', 'position', 'hire', 'find', 'mức lương', 'kinh nghiệm', 'tuyển', 'vị trí']
script = ''
posts = []
for i in range(0, 1440):
    for word in kw:
        post = data['content'].iloc[i]
        if word in post and post not in posts:
            posts.append(post)
            script += post
            data.drop(data.index[i])
            
```

```python
pd.set_option('display.max_colwidth', -1)
df = pd.DataFrame({'post':posts})
```

```python
sw=['làm', 'với', 'tới', 'liên', 'hệ', 'vui', 'lòng', 'nhiều', 'Việt', 'Nam', 'tham', 'gia', 'số','và', 'kinh', 'nghiệm', 'cơ', 'của', 'vào','không', 'sử', 'dụng', 'cũng', 'thì', 'nên', 'hội', 'yêu', 'cầu', 'như', 'có', 'để', 'cho', 'các',  'bạn', 'và', 'hấp', 'dẫn', 'đăng' ,'ký', 'về', 'and', 'là', 'một', 'là', 'khi', 'mà', 'được']
wordcloud = WordCloud(max_font_size=300, max_words=100, width = 5000, height=2500,background_color="black", stopwords = sw).generate(script)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
